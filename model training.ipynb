```python
def get_model():
    In = Input(shape=(32,32,1))
    x = Conv2D(32, (5,5), padding="same", activation="relu")(In)
    x = Conv2D(32, (5,5), activation="relu")(x)
    x = Conv2D(32, (5,5), activation="relu")(x)
    x = MaxPooling2D((2,2))(x)
    x = BatchNormalization()(x)
    
    x = Conv2D(64, (5,5), padding="same", activation="relu")(x)
    x = Conv2D(64, (5,5), activation="relu")(x)
    x = Conv2D(64, (5,5), activation="relu")(x)
    x = MaxPooling2D((2,2))(x)
    x = BatchNormalization()(x)
    
    x = Flatten()(x)
    x = Dense(128, activation="relu")(x)
    x = Dense(128, activation="relu")(x)
    x = Dropout(0.4)(x)
    
    Out = Dense(28, activation="softmax")(x)
    
    model = Model(In, Out)
    model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

model = get_model()
model.summary()
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import os
from tqdm import tqdm
```

```python
import os
from PIL import Image

image_dir = "dataset/train"
images = []
labels = []

for filename in tqdm(os.listdir(image_dir)):
    if filename.endswith(".png"):
        # Open the image and convert it to RGB format
        image = Image.open(os.path.join(image_dir, filename))
        # Resize the image to a fixed size (e.g. 224x224)
        image = image.resize((32,32))
        # Append the image and label to their respective lists
        images.append(np.array(image))
        labels.append(filename.split("_")[-1].split('.')[0]) # assuming the label is the first part of the filename before the underscore
df = pd.DataFrame(columns=[i for i in range(1024)])
for i in tqdm(range(len(images))):
    flat_img = list(images[i].ravel())
    df.loc[i] = flat_img
labels = [int(x)-1 for x in labels]
df = df/255
df['label'] = labels
```

```python
x = df.drop('label', axis = 1)
x = x.to_numpy().reshape(-1,32,32,1)
y = df['label']
```

```python
y = to_categorical(y)
```

```python
def get_model():
    In = Input(shape=(32,32,1))
    x = Conv2D(32, (5,5), padding="same", activation="relu")(In)
    x = Conv2D(32, (5,5), activation="relu")(x)
    x = Conv2D(32, (5,5), activation="relu")(x)
    x = MaxPooling2D((2,2))(x)
    x = BatchNormalization()(x)
    
    x = Conv2D(64, (5,5), padding="same", activation="relu")(x)
    x = Conv2D(64, (5,5), activation="relu")(x)
    x = Conv2D(64, (5,5), activation="relu")(x)
    x = MaxPooling2D((2,2))(x)
    x = BatchNormalization()(x)
    
    x = Flatten()(x)
    x = Dense(128, activation="relu")(x)
    x = Dense(128, activation="relu")(x)
    x = Dropout(0.4)(x)
    
    Out = Dense(28, activation="softmax")(x)
    
    model = Model(In, Out)
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

model = get_model()
model.summary()
```

```python
history = model.fit(x,y, epochs = 20, batch_size = 64, validation_split=0.2, verbose = 1, use_multiprocessing=True)
```

```python
import cv2
import numpy as np

def cartoonify(image_path):
    # Read the input image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply bilateral filter to remove noise and keep edges sharp
    gray_smooth = cv2.bilateralFilter(gray, 7, 75, 75)

    # Apply adaptive thresholding to create a binary image
    gray_edges = cv2.adaptiveThreshold(gray_smooth, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 2)

    # Combine the grayscale and binary images using bitwise operations
    color_edges = cv2.cvtColor(gray_edges, cv2.COLOR_GRAY2BGR)
    cartoon = cv2.bitwise_and(image, color_edges)

    return cartoon

input_path = "//home/waiz/Downloads//GettyImages-1092658864_hero-1024x575.jpg"
cartoon = cartoonify(input_path)
cv2.imshow("Cartoon", cartoon)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np

# Load the image
img = cv2.imread('/home/waiz/Downloads//GettyImages-1092658864_hero-1024x575.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply median blur to reduce noise
gray = cv2.medianBlur(gray, 5)

# Detect edges using adaptive thresholding
edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)

# Apply bilateral filter to smooth the image while preserving edges
filtered = cv2.bilateralFilter(img, 9, 75, 75)

# Apply color quantization using k-means clustering
Z = filtered.reshape((-1,3))
Z = np.float32(Z)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 20, 1.0)
K = 8
ret,label,center=cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
quantized = res.reshape((img.shape))

# Combine the edges and color quantized image to create the cartoonified image
cartoon = cv2.bitwise_and(quantized, quantized, mask=edges)

# Display the cartoonified image
cv2.imshow('Cartoonified Image', cartoon)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
